```python
import numpy as np

# Create a NumPy array of 5 numbers
numbers = np.array([5, 15, 25, 35, 45])

# Display the array
print("Numbers in the array â†’", numbers)

# Calculate sum and average using methods instead of np.sum/np.mean
total = numbers.sum()
average = numbers.mean()

# Display results neatly
print(f"\nTotal Sum = {total}")
print(f"Average Value = {average:.2f}")
```

```python
import matplotlib.pyplot as plt

# New Data
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]  # y = xÂ² pattern

# Plot line graph with a new style
plt.plot(x, y, color='red', marker='s', linestyle='--', linewidth=2, markersize=8)

# Add labels and title
plt.xlabel("X Values")
plt.ylabel("Y = XÂ²")
plt.title("Square Function Line Graph")

# Customize background and grid
plt.gca().set_facecolor("lavender")
plt.grid(True, linestyle=':', linewidth=1)

# Display the plot
plt.show()
```
