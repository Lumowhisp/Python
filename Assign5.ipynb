```python
import numpy as np

# Create an array of 5 numbers
arr = np.array([10, 20, 30, 40, 50])

# Print the array
print("Array:", arr)

# Calculate and print the sum
sum_arr = np.sum(arr)
print("Sum of array elements:", sum_arr)

# Calculate and print the average
avg_arr = np.mean(arr)
print("Average value of array elements:", avg_arr)
```

```python
import pandas as pd

# Create data using a dictionary
data = {
    'Name': ['Aman', 'Abhishek', 'Shivam', 'Aditya'],
    'Age': [20, 21, 19, 20]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print("DataFrame:\n")
print(df)
```

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

# Plot line graph
plt.plot(x, y, marker='o', color='blue', linestyle='-', linewidth=2)

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Line Graph")

# Show grid and display
plt.grid(True)
plt.show()
```
