```python
#Name:Aditya
#Roll:2401920100035
#Qtn_3
#partA:
#Wap to accept sentence from user and demonstrate length of the sentence ,slicing (first 5 char and last 3 char),conversion to uppercase and lowercase ,finding a substring enetred by user
# Accept a sentence
sentence = input("Enter a sentence: ")

# Length of the sentence
print("Length:", len(sentence))

# Slicing
print("First 5 characters:", sentence[:5])
print("Last 3 characters:", sentence[-3:])

# Uppercase and lowercase
print("Uppercase:", sentence.upper())
print("Lowercase:", sentence.lower())

# Finding substring
sub = input("Enter substring to find: ")
if sub in sentence:
    print("Substring found!")
else:
    print("Substring not found!")
```

```python
#Name:Aditya
#Roll:2401920100035
#Part C
# Define a tuple of 5 fruits
fruits = ("apple", "banana", "mango", "orange", "grapes")

# Indexing
print("First fruit:", fruits[0])      # apple
print("Third fruit:", fruits[2])      # mango
print("Last fruit:", fruits[-1])      # grapes

# Slicing
print("First three fruits:", fruits[0:3])   # ('apple', 'banana', 'mango')
print("Middle fruits:", fruits[1:4])        # ('banana', 'mango', 'orange')
print("All except first:", fruits[1:])      # ('banana', 'mango', 'orange', 'grapes')
print("All except last:", fruits[:-1])      # ('apple','banana','mango','orange')
```

```python
#WAP that uses Numpy to create s data frame containing at least 5 student marks in 3 subjects  and calculate and display the avg marks in each subject
import numpy as np
import pandas as pd

# Create a NumPy array (5 students Ã— 3 subjects)
marks = np.array([
    [85, 78, 92],
    [76, 88, 81],
    [90, 67, 79],
    [72, 95, 85],
    [88, 82, 90]
])

# Create DataFrame
df = pd.DataFrame(marks, columns=["Maths", "Physics", "Chemistry"])

print("Student Marks DataFrame:\n")
print(df)

# Calculate average marks of each subject
avg_marks = df.mean()

print("\nAverage Marks in Each Subject:\n")
print(avg_marks)
```
